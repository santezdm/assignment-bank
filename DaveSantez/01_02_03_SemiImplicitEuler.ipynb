# The Symplectic Euler Method with Regards to a Forced Harmonic Oscillator, and Other Semi-Implicit Method Tidbits

Throughout much of this course, we have used numerous numerical schemes utilizing Euler's integration method (forward, backward, central difference) to analyze phycial 1D and 2D systems of ODEs and PDEs. Over time, we came to discover a couple of problems that can arise when using these schemes, and a couple of things we have to work around to achieve stability of the scheme and convergence onto the solution. For example, we found that the numerical error is proportional to the step size (which can be a mess for some systems), CFL conditions are required (inherent of explicit integration schemes), Euler's method can have trouble with sinusoidal initial conditions, and so on. However, there was a reason we started with Euler's method as an introduction to numerical methods....because it is one of the simplest methods to understand, code, and serves as a stepping stone to more complex numerical methods [1] So, how can we improve Euler's method without truly overcomplicating the computation? This is where we will begin to study what's known as the Symplectic Euler method. 

To begin, let's look at a forced spring-mass-damper system, while ignoring friction.

[insert picture 1 here]

This setup can be described by the damped harmonic oscillator equation:

md2x/dt2+cdx/dt+kx = F

with initial conditions x(0) = a, x'(0) = b, where a and b are constants

What we have here is a second order, linear, ordinary differential equation. Solving this numerically would require solving two, separate, first-order ordinary differential equations with the Euler method thereby increasing numerical error due to two explicit integrations. Also, the size of the time step would be an issue as we would have to define a CFL, etc., etc.

However, we can use a comparably simple method called the Symplectic Euler Method (SEM), which is defined by its semi-implicit, semi-stable scheme. Remember the difference between implicit/explicit schemes?

**Explicit**: Uses equations describing the system during only the current time step to calculate the state of a system at a later time step.

**Implicit**: Uses equations describing the system during the current time step AND a later time to calculate the state of a system at a later time step.

So what does it mean for a method to be "semi-implicit" (and hence "semi-explicit") like the SEM? We'll delve into that a little later. For now, let's describe the SEM using some equations!

The SEM can be described as an application of a symplectic integrator used to solve a Hamiltonian system [2], which really just means that this method can be applied to some classical mechanical systems and some (very) specific quantum mechanical systems in a way that preserves the energy equation exactly. For our simple case, we can construct the following two first order, differential equations:

dx/dt = v(t)

dv/dt = -k/mx(t)-c/mv(t)+F/m

which, if we some common definitions inherent of a harmonic oscillator, namely:

damping ratio (m/s) = gamma = c/m

natural frequency^2 (rad/s)^2 = omega_o^2 = k/m

we end up with:

dx/dt = v(t)

dv/dt = -omega_o^2x(t)-gammav(t)+F/m

Using the SEM (essentially a forward time integration), we can get the governing equations:

x_n+1 = x_n + dt(v_n+1)

v_n+1 = v_n + dt(-omega_o^2x_n-gammav_n+F/m)

Here, in the calculation of x_n+1, we use the velocity of the mass in the next time step, v_n+1, hence the implicit-ness of the method. The traditional Euler method would instead calculate x_n+1 using v_n (the velocity at the current time step). However, we can already see from this set of equations that the velocity at the next time step is calculated *explicity*. This is the core difference between Euler's method and the SEM. Let's do a quick example to see some of the characteristics of the SEM.

```python
# Importing some libraries
import numpy as np                  
import matplotlib.pyplot as plt                 
%matplotlib inline
from matplotlib import rcParams
rcParams['font.family'] = 'serif'
rcParams['font.size'] = 16
```

### Initial Conditions of the Harmonic Oscillator system:

```python
m = 5.     # kg
c = 1.0     # N-s/m
k = 10.    # N/m
F_ext = 40. # N

omega = sqrt(k/m)     # (rad/s)^0.5
gamma = c/m           # m/s
```

### Defining our timeframe:

```python
T = 30.  # seconds
nt = 1000. # intervals
dt = T/nt
t = np.linspace(0,T,nt)
```

Now we get to define our SEM function, which describes the governing equations:

```python
def symplecticEuler(xn,vn,omega,gamma,F_ext,m,dt,nt):
    
#      Solves the simple harmonic oscillator system of equations using the Symplecitic Euler Method
#
#      Parameters
#      ----------
#      xn        : initial position of mass (m)
#      vn        : initial velocity of mass (m/s)
#      omega     : calculated natural frequency of the system ((rad/s)^0.5)
#      gamma     : calculated damping ratio of the system (m/s)
#      F_ext     : external force being applied to mass (N)
#      m         : mass of object (kg)
#      dt        : time step length (s)
#      nt        : number of time steps for simulation
#
#      Returns
#      -------
#      x         : array of positions of the mass
    
    x = xn.copy()
    for i in range(int(nt)):
        if i != 0:
            vn = v.copy()
            
        xn = x.copy()
        v = findV(xn,vn,omega,gamma,F_ext,m)
        x[1:] = xn[:-1]+dt*v[:-1]

    return x
```

We require the findV function to calculate the velocity of the mass at a later time step which can defined by the following:

```python
def findV(xn,vn,omega,gamma,F_ext,m):
        
#      Solves the simple harmonic oscillator system of equations using the Symplecitic Euler Method
#
#      Parameters
#      ----------
#      xn        : initial position of mass (m)
#      vn        : initial velocity of mass (m/s)
#      omega     : calculated natural frequency of the system ((rad/s)^0.5)
#      gamma     : calculated damping ratio of the system (m/s)
#      F_ext     : external force being applied to mass (N)
#      m         : mass of object (kg)
#
#      Returns
#      -------
#      v         : calculation of the velocity of the mass at the next time step

    v = vn.copy()
    v[1:] = vn[:-1]+dt*(-omega**2*xn[:-1]-gamma*vn[:-1]+1./m*F_ext)
    
    return v
```

The next function describes the same system using Euler's method:

```python
# Initializing the arrays

xn = np.zeros(nt)

vn = np.zeros(nt)
```

Now the moment of truth: some results!

```python
x_SEM = symplecticEuler(xn,vn,omega,gamma,F_ext,m,dt,nt)
plt.plot(t,x_SEM)
title('Forced Harmonic Oscillator Solved Using SEM')
xlabel('time elapsed (s)'); ylabel('x-displacement (m)')
plt.show()
```

We should also compare these results with the exact solution to the problem, which we can calculate by solving the characteristic equation of the harmonic oscillator:

r^2 + gamma(r)+ omega_o^2 = F/m

Using our initial conditions, (and after a little alegbra!) we will find that:

x(t) = -4exp(-t/10)sin(sqrt(199)t/10)/sqrt(199)-4exp(-t/10)cos(sqrt(199)t/10)+4

**(Note: If you forget how to solve this system of equations exactly, please refer to http://en.wikipedia.org/wiki/Harmonic_oscillator)**

Here, we'll plot our analytical solution with the exact solution:

```python
x_true = xn.copy()
for j in range(int(nt)):
    x_true[j] = -4.*exp(-t[j]/10.)*sin(sqrt(199)*t[j]/10.)/sqrt(199)-4.*exp(-t[j]/10.)*cos(sqrt(199)*t[j]/10.)+4
```

```python
plt.plot(t,x_SEM,'r',t,x_true,'g')
title('Forced Harmonic Oscillator SEM vs. Exact')
xlabel('time elapsed (s)'); ylabel('x-displacement (m)')
plt.legend(['SEM','Exact'],'upper right')
plt.show()
```

We can see in the plot above that, while the SEM model does capture the period of oscillation, there is still a distinct error as time elapses. The reason for this behavior will become clearer in a bit. Now, let's make sure the SEM model will converge on the solution. If we increase the time elapsed to, say, 100 s while retaining size of the time step (0.03 s), we'll get the following:

```python
T = 100.  # seconds
dt = 0.03

def SEMvsExact(dt,T):
    
    nt = T/dt
    t = np.linspace(0,T,nt)

    # Initializing the arrays
    xn = np.zeros(nt)
    vn = np.zeros(nt)

    x_SEM = symplecticEuler(xn,vn,omega,gamma,F_ext,m,dt,nt)
    x_true = xn.copy()
    for j in range(int(nt)):
        x_true[j] = -4.*exp(-t[j]/10.)*sin(sqrt(199)*t[j]/10.)/sqrt(199)-4.*exp(-t[j]/10.)*cos(sqrt(199)*t[j]/10.)+4
    
    plt.figure(1)
    plt.plot(t,x_SEM,'r',t,x_true,'g')
    title('Forced Harmonic Oscillator SEM vs. Exact')
    xlabel('time elapsed (s)'); ylabel('x-displacement (m)')
    plt.legend(['SEM','Exact'],'upper right')

SEMvsExact(dt,T)
```

Perfect! The estimated solution converges to the exact solution after a large period of time. We can say (and others have proved!), because these equations describe Hamiltonian motion, that the SEM *almost* conserves energy. It's said the energy of the same system would steadily increase when using the traditional Euler method, making potential solutions even less accurate. [2] This concept of conserving energy is very important when discussing mechanical systems. Now, let's see how the semi-implicit/explicit nature of the method affects our calculated solution.

### Semi-Implicit Method Characteristics

Let's take a look at the same problem, with say time steps of 0.1 s. If the SEM was fully implicit, we could choose fairly large time steps and still get acceptable results.

```python
T = 100.  # seconds
dt = 0.1

SEMvsExact(dt,T)
```

Whoa, that looks bad. If you exteded the elapsed time, the solution would never converge on the exact solution! While we're at it, let's see what happens when some system parameters change (i.e., damping ratio and natural frequency). For now, let's go back to the "good" time step of 0.03 s and change just the spring constant to 100 N/m...a feasible spring constant in some machinery to be certain.

```python
m = 5.     # kg
c = 1.0     # N-s/m
k = 100.    # N/m
F_ext = 40. # N

omega = sqrt(k/m)     # (rad/s)^0.5
gamma = c/m           # m/s

T = 100.  # seconds
dt = 0.03

SEMvsExact(dt,T)
```

Well, this doesn't look too good either! It turns out, because the SEM is considered semi-implicit, there exists a certain stability region where the solution will be correctly simulated when using the SEM. The complex roots of the characteristic equation must lie within this region for the integrated solution to be stable, and there does exist boundaries that are akin to the CFL conditions in Euler's method. However, upon further examination of how the stability region was discovered (refer to Niiranen's work! [4]), it can be seen that the stability region of the traditional Euler method is much smaller than the SEM stability region. [2]
